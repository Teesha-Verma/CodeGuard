# Jupyter Notebook 2 — Complexity Analysis

**Purpose:** Measure Cyclomatic Complexity and Maintainability Index using Radon. Map findings to selective thresholds (CC > 10, MI < 65).

```python
from radon.visitors import ComplexityVisitor
from radon.metrics import mi_visit

complex_code = """
def nested_calculator(a, b, c):
    if a > 0:
        if b > 0:
            if c > 0:
                return a + b + c
            else:
                return a + b
        else:
            return a
    return 0
"""

visitor = ComplexityVisitor.from_code(complex_code)
for func in visitor.functions:
    print(f"Function: {func.name} | Complexity: {func.complexity} (Rank: {func.rank})")

mi = mi_visit(complex_code, multi=True)
print(f"Maintainability Index: {mi}")
```
