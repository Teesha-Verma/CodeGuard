# Jupyter Notebook 1 — AST Experiments

**Purpose:** Inspect Python syntax structures deeply, test custom visitor logic, and verify AST node extraction for classes, functions, and control structures.

```python
import ast

test_code = """
class UserAuth:
    def login(self, username, password):
        if username == "admin":
            eval(password)  # Dangerous Call
        return False
"""

tree = ast.parse(test_code)
print(ast.dump(tree, indent=2))
```

### Function and Class Extraction Experiment

```python
class CustomVisitor(ast.NodeVisitor):
    def visit_FunctionDef(self, node):
        print(f"Found Function: {node.name} at line {node.lineno} (args: {[a.arg for a in node.args.args]})")
        self.generic_visit(node)

    def visit_ClassDef(self, node):
        print(f"Found Class: {node.name} inherits {[b.id for b in node.bases if isinstance(b, ast.Name)]}")
        self.generic_visit(node)

CustomVisitor().visit(tree)
```
